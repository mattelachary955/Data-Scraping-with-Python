```python
import requests
from bs4 import BeautifulSoup as bs
from selenium import webdriver
import time
from tqdm import tqdm
import pandas as pd
import os
from shutil import copyfile as copy
```

```python
browser = webdriver.Edge()
```

```python
browser.get("https://stock-pictures.netlify.app")
```

```python
for i in tqdm(range(285)):
    browser.execute_script("window.scrollTo(" + str(i*1000) +","+ str((i+1)*1000) +")")
    time.sleep(0.15)
```

```python
soup = bs(browser.page_source, "html.parser")
```

```python
soup.text
```

```python
images_list = []
for sp in soup.find_all("div", class_ ="container"):
    image_link = sp.find("img").get("src")
    tags = ",".join(sp.find("span",class_ = "tag-color").text[7:].split(","))
    likes = sp.find("div", class_ = "likes-comments").text.replace("Likes","").replace("Comments", "").split(" ")[0]
    comments = sp.find("div", class_ = "likes-comments").text.replace("Likes","").replace("Comments", "").split(" ")[1]
    images_list.append([tags, likes, comments, image_link])
```

```python
len(images_list)
```

```python
df = pd.DataFrame(images_list, columns = ["tags","likes","comments","image_link" ])
```

```python
df.to_csv("Images.csv")
```

```python
df["image_link"]
```

```python
ref = 10000
count = 1
for link in tqdm(df["image_link"]):
    res = requests.get(link)
    path = "Images/" +"A"+str(ref + count)[1:] + ".jpg"
    with open(path, "wb") as file:
        file.write(res.content)
        file.close()
        count += 1
```

```python
df1["file_name"] = os.listdir("Images")
```

## Creating unique tags for all images

```python
tags = []
for tag in df1["tags"]:
    tags += str(tag).split(", ")
tags = list(set(tags))    
```

```python
tags
```

## Creating folder for each tag

```python
for tag in tags:
    try:
        os.mkdir("Classified Images/" + tag)
    except:
        pass
```

## Segregating images on folder based on tags 

```python
for data in tqdm(df1.values):
    tags = str(data[1]).split(", ")
    source = "Images/" + data[5]
    for tag in tags:
        dest = "Classified Images/" + tag + "/" + source.split("/")[-1]
        try:
            copy(source, dest)
        except:
            pass
```

## TOP 10 tags with highest number of images

```python
folders = []
freq = []
for folder in os.listdir("Classified Images"):
    freq.append(len(os.listdir("Classified Images/" + folder)))
    folders.append(folder)
```

```python
freq_df = pd.DataFrame()
```

```python
freq_df["folders"] = folders
freq_df["freq"] = freq
```
