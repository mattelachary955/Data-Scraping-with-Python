```python
import requests 
from bs4 import BeautifulSoup as bs
```

## ask your title here

```python
title = str(input("Enter your title: ")).replace(" ","+")
link = "https://www.google.com/search?q=" + title + "+wikipedia"
link
```

```python
res = requests.get(link)
soup = bs(res.text, "html.parser")
for sp in soup.find_all("div"):
    try:
        link = sp.find("a").get("href")
        if "en.wikipedia.org" in link:
            link = (link[7:].split("&")[0])
            break
    except:
        pass
print(link)
```

```python
res = requests.get(link)
soup = bs(res.text)
```

```python
heading = soup.find("h1", class_ = "firstHeading mw-first-heading").text
```

```python
heading
```

```python
para = ""
for result in soup.find_all("p"):
    para = para + (result.text) + "\n"  
```

```python
para
```

```python
for i in range(2,476):
    para = para.replace("[" + str(i) + "]", "")
```

```python
with open(heading + ".txt","w", encoding="utf-8") as f:
    f.write(para)
    f.close()
```
